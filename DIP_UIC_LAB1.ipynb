# 2)
Consider the moving average system h=1./5.*np.array([1.,1.,1.,1.,1.]).

If the input signal is x, then the output signal will be y=np.convolve(x,h).

For the following input signal x, print the graphic of the output.
(a) x=1  for k=[0:20]

(b). x=cos(0.1k) for k=[-40:80]

(c). x=cos(0.95k) for k=[-40:80]

(d). x is the signal obtained in Part A.3




```python
# define the filter (convolution  )
h=1./5. * np.array([1.,1.,1.,1.,1.])
print(h)
```

```python
# (a) x=1 for k=[0:20]
# define the signal
signal1 = np.zeros(shape=20)

for n in range(20):
  k=n
  signal1[n] = 1

#or simply
signal1 = np.ones(shape=20)
```

```python
#(b). x=cos(0.1k) for k=[-40:80]
# define the signal
signal2 = np.zeros(shape=120)

for n in range(120):
  k=n-40
  signal2[n] = np.cos(k*0.1)
```

```python
#x=cos(0.95k) for k=[-40:80]
# define the signal
signal3 = np.zeros(shape=120)

for n in range(120):
  k=n-40
  signal3[n] = np.cos(k*0.95)
```

```python
#d)x is the signal obtained in Part A.3  ====>    x[n]=7cos(0.1n)+cos(0.95n)
# define the signal
signal4 = np.zeros(shape=120)

for n in range(120):
  k=n-40
  signal4[n] = 7 * np.cos(k*0.1) + np.cos(k*0.95)
```

```python
#signal = signal0
#signal = signal1
#signal = signal2
#signal = signal3
signal = signal4



y=np.convolve(signal,h)
fig5 = plt.figure(5)
plt.stem(signal)
plt.title("Signal")
#plt.show()
fig6 = plt.figure(6)

plt.stem(y)
plt.title("Convole")

plt.show()
```
